```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create the basic structure of the DF
data = pd.DataFrame(np.nan,
                    index=np.arange(27*3),
                    columns=["replicate",
                             "cell_cell", "cell_neurite", "neurite_neurite",
                             "rosettes", "partial", "linear", "empty"])

# Apply the conditions for each of the simulations
data["replicate"] = np.tile([1, 2, 3], 27)

data["cell_cell"] = np.tile(np.array([np.repeat("Low", 3), 
                                      np.repeat("Balance", 3), 
                                      np.repeat("High", 3)]).flatten(), 9)

data["cell_neurite"] = np.tile(np.array([np.repeat("Low", 9), 
                                         np.repeat("Balance", 9), 
                                         np.repeat("High", 9)]).flatten(), 3)

data["neurite_neurite"] = np.tile(np.array([np.repeat("Low", 27), 
                                            np.repeat("Balance", 27), 
                                            np.repeat("High", 27)]).flatten(), 1)

#data.to_csv("rosette_count.csv", index=False)
```

```python
data = pd.read_csv("rosette_count.csv")

low_nn = data[data["neurite_neurite"] == "Low"]
bal_nn = data[data["neurite_neurite"] == "Balance"]
high_nn = data[data["neurite_neurite"] == "High"]
```

```python
fig, axes = plt.subplots(1, 5, figsize=(13, 3), gridspec_kw={'width_ratios':[1,1,1,1,0.08]})
vmin = 0
vmax = 3

for i, rose_type in enumerate(["rosettes", "partial", "linear", "empty"]):
    ax = axes[i]
    piv = pd.pivot_table(low_nn, values=rose_type, index=["cell_cell"], columns=["cell_neurite"])
    piv = piv.reindex(columns=["Low", "Balance", "High"], index=["Low", "Balance", "High"])

    sns.heatmap(data=piv, cmap="Blues", square=True, annot=True,
                linewidth=1, ax=ax, cbar=i==0, cbar_ax=axes[-1],
                vmin=vmin, vmax=vmax)
    ax.invert_yaxis()
    ax.set_title(f"Structure: {rose_type}")
    if i > 0:
        ax.set_ylabel("")
        ax.set_yticks([])
    else:
        ax.set_ylabel("Cell-cell repulsion", labelpad=15, fontsize=13)
    ax.set_xlabel("Cell-neurite repulsion", labelpad=15, fontsize=13)
    
fig.suptitle("Mean number of different spatial structures in simulations.", y=1.05, fontsize=14)
plt.savefig("Structures.png")
```
